```
%cd ml/ml-latest-small
rf=pd.read_csv('ratings.csv')
mf=pd.read_csv('movies.csv')
```

```

df=mf.merge(rf,on='movieId')

df.drop(['genres','timestamp'],axis=1,inplace=True)
df.head()
```

```
rf=df.groupby('movieId')['movieId','rating','title'].mean()
rf['count']=df.groupby('movieId')['rating'].count()
rf=rf.sort_values('count',ascending=False)
mn=pd.DataFrame(df.groupby('movieId')['title'],columns=['movieId','title'])
rf.index.name='mid'
rf=rf.merge(mn,on='movieId')
rf.tail()
```

```
rf[rf['count']>100]
df=df[df['movieId'].isin(rf[rf['count']>100]['movieId'])]
df.head()
```

```
mat=df.pivot_table(index='userId',columns='movieId',values='rating')
```
